Birleştirmeli Sıralama (Merge Sort) ve İkili Arama (Binary Search) Uygulaması

1. Dizi Tanımlama

```python
array = [98, 17, 44, 35, 94]
```

2. Birleştirmeli Sıralama (Merge Sort)

```python
def mergeSort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left = arr[:mid]
        right = arr[mid:]

        mergeSort(left) # Solu bir elemana kadar bol
        mergeSort(right) # Sagi bir elemana kadar bol

        i = j = k = 0

        # Sol ve sagda bir eleman kalıncaya kadar kostur
        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                arr[k] = left[i]
                i += 1
            else:
                arr[k] = right[j]
                j += 1
            k += 1

        # Sol tek eleman kalmissa
        while i < len(left):
            arr[k] = left[i]
            i += 1
            k += 1
        
        # Sag tek eleman kalmissa
        while j < len(right):
            arr[k] = right[j]
            j += 1
            k += 1

    return arr
```

3. Diziyi Sırala

>- Sırasız dizi:

```python
print("Dizi: ", array)
```

>- Sıralanmış dizi (newArr 'e sıralanmış diziyi kaydet)

```python
newArr = mergeSort(array)
print("Siralanmis Dizi: ",newArr)
```

4. İkili Arama (Binary Search)

```python
def binarySearch(arr, target):
    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1
```

5. Aramayı Gerçekleştirme

>Sıralanmış dizi içerisinde, value elemanını arama

```python
value = 44 # Aranacak deger
index = binarySearch(arr=newArr, target=value)

if index == -1:
    print("Eleman bulunamadi: ", value)
else:
    print("Eleman bulundu: ", value, "\nIndeksi:", index)
```

- Binary Search algoritması sürekli diziyi bölüp, küçük dizi(cik)ler içerisinde arananı bulmaya çalışır.
- Big-O analizi O(logn) dir.

>8 Elemanlı bir dizi için, 

>>log8 = 3 değerini verir, yani 8 elemanlı dizide bir elemanın bulunması maksimum 3 bolmeden sonra tamamlanır.

- Merge Sort algoritması da sürekli diziyi tek elemanlı birer dizi(cik)ler olana kadar bölmeye devam eder.
- Binary Search gibi Big-O analizi, bu durum için O(logn) olur. Ancak:

>Merge aşamasında her seviyede toplam n işlem yapılmaktadır.

>>Toplam logn seviye olduğu için MergeSort'un Big-O analizi, O(nlogn) olacaktır.
